```python
url = 'http://ergast.com/api/f1/2008/5/driverStandings.json'
```

```python
#response object()
response = requests.get(url)
print(response)
```

```python
# We only want the Driver Standings
for x in range(5):
    my_racer_data = response.json()['MRData']['StandingsTable']['StandingsLists'][0]['DriverStandings'][x]['Driver']['familyName']
    print(my_racer_data)
```

```python
def get_racer_info(data):
    new_data = []
    for racer in data:
        racer_dict = {}
        racer_name = f'{racer["Driver"]["givenName"]} {racer["Driver"]["familyName"]}'
        
        racer_dict[racer_name]={
            'first_name':racer['Driver']['givenName'],
            'last_name':racer['Driver']['familyName'],
            'position':racer['position'],
            'wins':racer['wins'],
            'DOB':racer['Driver']['dateOfBirth'],
            'nationality':racer['Constructors'][0]['name']
        }
        new_data.append(racer_dict)
    return new_data

get_racer_info(my_racer_data)
```

```python
def get_driver_info_by_year_and_round(year, r):
    url = f"https://ergast.com/api/f1/{year}/{r}/driverStandings.json"
    response = requests.get(url)
    if not response.json()["MRData"]["StandingsTable"]["StandingsLists"]:
        return "We had an error loading your data. It is likely the year or round is not in the database."
    data = response.json()["MRData"]["StandingsTable"]["StandingsLists"][0]["DriverStandings"]
    return get_racer_info(data)

get_driver_info_by_year_and_round(2020, 4)
```

```python
import requests, json

def get_pokemon_info(pokemon_name):
    response = requests.get(f'https://pokeapi.co/api/v2/pokemon/{pokemon_name}/')
    if response.status_code == 200:
        pokemon_data = response.json()
        return {
            'abilities': [ability['ability']['name'] for ability in pokemon_data['abilities']],
            'types': [poke_type['type']['name'] for poke_type in pokemon_data['types']],
            'weight': pokemon_data['weight']
        }

def make_pokemon_dict(pokemon_names):
    pokemon_dict = {}
    for name in pokemon_names:
        data = get_pokemon_info(name)
        if data:
            pokemon_dict[name] = data
    return pokemon_dict

def categorize_by_type(pokemon_dict):
    categorized_dict = {}
    for name, data in pokemon_dict.items():
        for poke_type in data['types']:
            if poke_type not in categorized_dict:
                categorized_dict[poke_type] = {}
            categorized_dict[poke_type][name] = data
    return categorized_dict

pokemon_names = ['pikachu', 'blastoise', 'haunter', 'dragonite', 'ditto', 'eevee', 'charizard', 'mimikyu', 'jigglypuff', 'mewtwo', 'slowpoke', 'togepi', 'bulbasaur', 'cubone', 'snorlax', 'ninetails', 'jolteon', 'vaporeon']

pokemon_dict = make_pokemon_dict(pokemon_names)

categorized_pokemon = categorize_by_type(pokemon_dict)

print(categorized_pokemon)
```
